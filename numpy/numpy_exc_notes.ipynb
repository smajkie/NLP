# Statistical operations

**The median** is a measure of central tendency that represents the middle value in a sorted list of numbers. To find the median, you first need to arrange the numbers in ascending or descending order. If the list has an odd number of observations, the median is the number that is exactly in the middle of the list. If the list has an even number of observations, the median is the average of the two middle numbers.

**Standard deviation** is a statistical measure that quantifies the amount of variation or dispersion in a set of values. It is a widely used measure of the uncertainty or spread in a set of data. The standard deviation indicates how much the individual data points in a data set deviate from the mean (average) value of the set.

**Variance** is a statistical measure that represents the degree of spread in a set of data points. More specifically, it quantifies the average squared deviation of each number from the mean of the data set. Variance provides a way to quantify how much the numbers in the data set differ from each other and from the mean.

```python
print(np.mean(d2), d2.mean())                                                   # dwa spososby
print(np.std(d2), d2.std())
print(np.var(d2), d2.var())
print(np.min(d2), d2.max())                                                     #
print(np.argmin(d2), d2.argmax())                                               # wszystkie te przyklady robia to na calej macierzy bez wzgledu na jej wymiar czyli biorÄ… wsztstkie jej elementy

print(d2.mean(axis=0))                                                          # sredna po kolumnach
print(d2.mean(axis=1))                                                          # srednia po wierszach
```

# Reshaping arrays

In NumPy, the ***`shape`*** attribute of an array tells you the dimensions of the array (i.e., how many rows, columns, and so on), while the ***`reshape`*** method allows you to change the dimensions of the array without changing the data within it.

In NumPy, the ***`ravel`*** and ***`flatten`*** methods are used to convert a multi-dimensional array into a one-dimensional array. Both methods accomplish similar goals, but there are some differences in how they work and the results they produce.

The key difference between ravel and flatten is that ravel will return a view of the original array whenever possible (this means that changes to the result of ravel may also affect the original array), whereas flatten will always return a copy (changes to the result of flatten will not affect the original array).

```python
print(d3.shape)
print(d3)
print(d3.reshape(4,5))                                                          # przeksztalcamy 3 wymiarowa w 2 wymiarowa, 4 wiersze 5 kolumn
print(d3.ravel())                                                               # przeksztalcamy 3 wymiarowa w 1 wymiarowa przeksztalcajac oryginal
print(d3.shape)
print(d3.flatten())                                                             # przeksztalcamy 3 wymiarowa w 1 wymiarowa tworzac nowa wynikowa i nie dotykajac oryginalu
print(d3.shape)
print(d3)

```

# Sequence, repetitions, random numbers

```python
a15 = np.arange(1,20,2)
print(a15)
a16 = np.repeat(a15,3)                                                          # repeat powiela kazdy element x razy i dalej nastepny element x razy itp, repeat robi tez automatycznie flatening ale mozna podac axis i wtedy zrobi jak trzeba
print(a16)
a17 = np.tile(a15,3)                                                            # tile powiela cala tablice x razy
print(a17)
from numpy import random
a18 = random.random((3,3,3))
print(a18)
a19 = random.randint(1,10, size=(3,3,3))                                         # od do ile elementow jaki rozmiar itp
print(a19)
```

# Where, argmax argmin

```python
a20 = np.where(a19>3, a19, 0)
print(a20)
a20 = np.where(a19>3, a19, np.random.randint(100,110, size=(3,3,3)))
print(a20)
```

```python
print(a20.max(), np.argmax(a20))
print(a20[:, 0])                                                # tylko pierwszy
print(a20[0, :])                                                # tylko pierwszy wiersz
```

Concatenate, vstack, hstack, sort, argsort

```python
b1 = np.array([1,2,3,4,5,6])
b2 = np.array([7,8,9,10,11,12])
b3 = np.concatenate((b1,b2))
b4 = np.vstack((b1,b2))
b5 = np.hstack((b1,b2))
print(b3, b5)
print(b4)

np.random.shuffle(b3)
print(b3)
b3 = np.sort(b3)
print(b3)


```
